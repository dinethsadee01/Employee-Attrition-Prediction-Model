### **1. Logistic Regression**

  * <b>Strengths:</b>
    - Easy to implement and interpret.

  * <b>Weaknesses:</b>

    - Performance may drop with non-linearly seperated data.

  * <b>Applicability:</b>
    - Suitable for binary classification and small to medium-sized datasets.
  
  * <b>Why it could work:</b>
    - Provides a baseline for comparison and interpretable coefficients for feature importance.

  * <b>Strengths:</b>

    - Provides feature importance scores.

  * <b>Weaknesses:</b>
    - Requires more computational resources than simpler models.
    - Harder to interpret compared to Logistic Regression.

  * <b>Applicability:</b>
    - Effective for large datasets with mixed feature types.

  * <b>Why it could work:</b>
    - Captures complex patterns in the data and performs well on tabular datasets.

  * <b>Strengths:</b>
    - Highly efficient and optimized for large datasets.
    - Handles imbalanced datasets using inbuilt mechanisms.
    - Flexible and offers hyperparameter tuning options.

  * <b>Weaknesses:</b>
    - Requires careful hyperparameter tuning to avoid overfitting
    - Computationally expensive compared to simpler models.
  * <b>Applicability:</b> Suitable for large, complex datasets where feature interactions are important.

  * <b>Why it could work:</b>
    - Already proven effective during implementation and achieves high performance metrics.

## Outlier Removing

```python
# Outlier emoval using IQR method
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

cleaned_df = remove_outliers(df, columns=['Monthly Income', 'Years at Company'])
```

## Algorithm Selection

Created some user defined functions for find the best,


*   Scaler
*   Train-test split
* Algorithm

User defined functions are,



1.   **`evaluate_models`** - taking x, y, candidate models, candidate splits and candidate scalers as parameters and return the performance of all the combinations as a data frame.
2.   **`visualize_results`** - taking the above models output as parameter then print, plot and save csv of the all results.
3. **`find_best_combination`** - taking the above models output csv and calculate weighted score of all the combinations and output the bst combination of scaler, split and algorithm to use.





### Try different scalers, splits and models

```python
# Model evaluation
models = {
    'DecisionTree': DecisionTreeClassifier(),
    'LogisticRegression': LogisticRegression(),
    'RandomForest': RandomForestClassifier(),
    'XGBoost': XGBClassifier()
}
scalers = [StandardScaler(), MinMaxScaler(), RobustScaler()]
splits = [0.7, 0.75, 0.8, 0.82, 0.83, 0.84, 0.85, 0.88, 0.9]

# Evaluation function defining
def evaluate_models(x, y, models, splits, scalers):
    all_results = []

    for scaler in scalers:
        # Scale features
        scaler_name = scaler.__class__.__name__
        x_scaled = scaler.fit_transform(x)
        for split in splits:
            x_train, x_test, y_train, y_test = train_test_split(
                x_scaled, y, test_size=(1-split), random_state=42
            )
            # Train the model
            for model_name, model in models.items():
                model.fit(x_train, y_train)
                y_pred = model.predict(x_test)
                y_proba = model.predict_proba(x_test)[:, 1] if hasattr(model, "predict_proba") else None

                # Calculate metrics
                metrics = {
                    "Scaler": scaler_name,
                    "Split": f"{round(split*100)}-{round((1-split)*100)}",
                    "Model": model_name,
                    "Accuracy": accuracy_score(y_test, y_pred),
                    "Precision": precision_score(y_test, y_pred, zero_division=0),
                    "Recall": recall_score(y_test, y_pred),
                    "F1-Score": f1_score(y_test, y_pred),
                    "ROC-AUC": roc_auc_score(y_test, y_proba) if y_proba is not None else "N/A"
                }
                all_results.append(metrics)

    return pd.DataFrame(all_results)

# Passing parameters to the function
results = evaluate_models(x, y, models, splits, scalers)
```

```python
# Visualizationfunction
def visualize_results(results):
    for metric in ["Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"]:
      g = sns.catplot(
          data=results,
          x="Split",
          y=metric,
          hue="Model",
          col="Scaler",
          kind="bar",
          height=5,
          aspect=1.5,
          ci=None
          )
      # Add titles and labels
      g.fig.subplots_adjust(top=0.85)
      g.fig.suptitle(f"Model Performance Comparison by Scaler ({metric})", fontsize=16)
      g.set_axis_labels("Train-Test Split Ratio", metric)
      g.set_titles("{col_name} Scaler")
      g.set_xticklabels(rotation=45)

    plt.show()

# Display, visualize and saving the results
print(results)
visualize_results(results)
results.to_csv('model_comparison_results.csv', index=False)
```

### Finding the best combination of scaler, split and model

```python
# Load the results CSV
results = pd.read_csv("model_comparison_results.csv")

# Define a function to rank by metrics and get the best combination
def find_best_combination(results):
    weights = {"Accuracy": 1, "Precision": 1, "Recall": 1, "F1-Score": 1, "ROC-AUC": 2} # Define weights for metrics
    for metric in weights.keys():
        results[metric + "_Norm"] = (results[metric] - results[metric].min()) / (results[metric].max() - results[metric].min()) # Normalize each metric between 0 and 1
    results["Score"] = sum(results[metric + "_Norm"] * weight for metric, weight in weights.items())# Compute a weighted score
    best_combination = results.sort_values("Score", ascending=False).iloc[0] # Sort by the highest score
    return best_combination

# Passing parameter to the function
best_combination = find_best_combination(results)

# Visualizing the best combinations
print("Best Combination:")
print(best_combination)
#results.to_csv('Best_Combinations.csv', index=False)
```

### Find best Hyper-Parameters

Hyperparameters are some parameters we can pass into the algorithms that allows to tweak model performance for optimal results.

```python
# Hiperparameter space for evaluation
space = {
    'n_estimators': hp.quniform('n_estimators', 50, 500, 10),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
    'max_depth': hp.quniform('max_depth', 3, 10, 1),
    'subsample': hp.uniform('subsample', 0.5, 1.0),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1.0),
    'gamma': hp.uniform('gamma', 0, 5),
    'reg_alpha': hp.uniform('reg_alpha', 0, 10)
}
```

1. **HYPEROPT** is a powerful python library that search through an hyperparameter space of values and find the best possible values that gives the minimum of the loss function.

2. **`Space`** is the input values over which we want to search.

3. Used hyperopt optimization algorithms are:


  *   **`hp.uniform(label, low, high)`** — Returns a value uniformly between low and high.
  *   **`hp.quniform(label, low, high, q)`** — Returns a value round(uniform(low, high) / q) * q, i.e it rounds the decimal values and returns an integer.

4. **`objective`** function is a function which returns a real value that we want to minimize. In this case, the real value is accuracy, then we want to maximize it. Then the function should return the negative of that metric.

(https://www.kaggle.com/code/prashant111/a-guide-on-xgboost-hyperparameters-tuning#A-Guide-on-XGBoost-hyperparameters-tuning)


```python
# Function to evaluate hyperparameters
def objective(space):
    model = XGBClassifier(
                    n_estimators = int(space['n_estimators']),
                    learning_rate = space['max_depth'],
                    max_depth = int(space['max_depth']),
                    subsample = space['subsample'],
                    colsample_bytree=int(space['colsample_bytree']),
                    gamma = space['gamma'],
                    reg_alpha = space['reg_alpha']
                    )

    evaluation = [( x_train, y_train), ( x_test, y_test)]

    model.fit(x_train, y_train,
            eval_set = evaluation, verbose=False)


    pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, pred>0.5)
    print ("Score:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
np.random.seed(42)
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 100,
                        trials = trials,
                        rstate = np.random.default_rng(42)
                        )

print("The best hyperparameters are : ","\n")
print(best_hyperparams)
```



*   Here best_hyperparams gives us the optimal parameters that best fit model and better loss function value.
*   `Trials` is an object that contains or stores all the relevant information such as hyperparameter, loss-functions for each set of parameters that the model has been trained.
* `‘fmin’` is an optimization function that minimizes the loss function and takes in 4 inputs - `fn, space, algo` and `max_evals`.
*   Algorithm used is `tpe.suggest`.

## Visualization of Result

```python
# 1. Confusion Matrix Visualization
def plot_confusion_matrix(y_test, y_pred, model_name):
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Left', 'Stayed'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()
    plt.tight_layout(pad=5.0)

# 2. ROC Curve
def plot_roc_curve(y_test, y_proba, model_name):
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = roc_auc_score(y_test, y_proba)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc='lower right')
    plt.show()
    plt.tight_layout(pad=5.0)

# 3. Feature Importance
def plot_feature_importance(model, feature_names):
    if hasattr(model, "feature_importances_"):
        importances = model.feature_importances_
        sorted_idx = importances.argsort()
        plt.figure(figsize=(10, 8))
        plt.barh([feature_names[i] for i in sorted_idx], importances[sorted_idx], color='skyblue')
        plt.title('Feature Importance')
        plt.xlabel('Importance Score')
        plt.ylabel('Features')
        for i, v in enumerate(importances[sorted_idx]):
            plt.text(v, i, f"{v:.4f}", color='black', va='center')
        plt.show()
    else:
        print("Feature importance not available for this model.")


# `X_test`, `y_test` are test data and `y_pred`, `y_proba` are predictions and probabilities
model_name = "XGBoost"
y_pred = loaded_model.predict(x_test)
y_proba = loaded_model.predict_proba(x_test)[:, 1]  # Probability for positive class

# Confusion Matrix
plot_confusion_matrix(y_test, y_pred, model_name)

# ROC Curve
plot_roc_curve(y_test, y_proba, model_name)

# Feature Importance
feature_names = x.columns
plot_feature_importance(loaded_model, feature_names)
```

**Observation**

1.   *Confusion Matrix*: to describe the performance of classification model. It demonstrates fairly balanced precision, recall, and specificity, which indicates it performs well in identifying both positive and negative cases.
* TP: 4500
* FN: 1553
* FP: 1487
* TN: 5057
2.   *ROC-AUC Curve*: an AUC of 0.85 means that, on average, this model ranks a randomly chosen positive instance higher than a randomly chosen negative instance 85% of the time. AUC values range from 0.5 (random guessing) to 1.0 (perfect classification).
3. *Feature Importance*: The most influential feature is Marital Status_Single, indicating that whether an employee is single has a significant impact on predicting attrition.Job Level and Remote Work are the next most important features.
